```python
import pandas as pd
```

```python
df = pd.read_excel('maca.xlsx', dtype='string')
df = df.apply(lambda x: x.str.strip())
```

```python
def pica_schreiben(row):

    pica = f"""0500 Af
0501 $btxt
0502 $bn
0503 $bnc
0551 X$bm
0600 ra;lf
1100 {row.Jahr}$n[{row.Jahr}]
1500 /1ger
2000 Broschur
2105 95,P01
4000 !550797866!
3010 !120883163!$4aut
4004 *{row.Zählung}.*{row.Titel} / Dan Shocker
"""
    if pd.notna(row.Auflage):
        pica +=f"4020 {row.Auflage}\n" 
           
    pica +=f"""4060 64 Seiten
4700 |FE-L|
7001 x
4800 !009030115!
7100 {row.Signatur} - {row.Zählung}
"""
    
    pica +="\t\n"

    return pica
```

```python
for index, row in df.iterrows():
    print(pica_schreiben(row))
```
